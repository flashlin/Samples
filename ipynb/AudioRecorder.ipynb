```python
import pyaudio
import wave

def record_audio(filename="tmp.wav", record_seconds=5):
    
    # 設定錄音參數
    FORMAT = pyaudio.paInt16  # 錄音的格式
    CHANNELS = 1  # 單聲道
    RATE = 16000  # 採樣率
    CHUNK = 1024  # 資料塊大小
    RECORD_SECONDS = record_seconds  # 錄音時間，單位為秒
    WAVE_OUTPUT_FILENAME = filename

    audio = pyaudio.PyAudio()

    # 開啟音訊串流
    try:
        stream = audio.open(
            format=FORMAT,
            channels=CHANNELS,
            rate=RATE,
            input=True,
            frames_per_buffer=CHUNK,
            input_device_index=3  # 設定錄音設備的索引
        )
    except Exception as e:
        print(f"無法開啟音訊串流: {e}")
        return

    frames = []  # 存放錄音資料的陣列

    # 錄音主迴圈
    for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
        data = stream.read(CHUNK)
        frames.append(data)

    # 關閉和釋放資源
    stream.stop_stream()
    stream.close()
    audio.terminate()

    # 儲存錄音檔案
    waveFile = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
    waveFile.setnchannels(CHANNELS)
    waveFile.setsampwidth(audio.get_sample_size(FORMAT))
    waveFile.setframerate(RATE)
    waveFile.writeframes(b''.join(frames))
    waveFile.close()
```

```python
import whisper
model = whisper.load_model("tiny")
```

## 錄製並轉錄聲音

```python
record_audio(filename="tmp.wav", record_seconds=5)
```

```python
from IPython.display import Audio
Audio("tmp.wav")
```

```python
audio = whisper.pad_or_trim(whisper.load_audio('tmp.wav'))
whisper.transcribe(model, audio)['text']
```

## 即時轉錄聲音

```python
while True:
    record_audio(filename="tmp.wav", record_seconds=5)
    audio = whisper.pad_or_trim(whisper.load_audio('tmp.wav'))
    print(whisper.transcribe(model, audio, fp16=False)['text'])
```
